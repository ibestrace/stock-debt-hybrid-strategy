```python
import pandas as pd
from IPython.display import display

stock = pd.read_excel('data/sh000905.xlsx',
                      index_col='date',
                      parse_dates=True)
returns = stock.pct_change().fillna(0)
returns = pd.Series(returns.close)
display(returns)
```

```python
%matplotlib inline
import quantstats as qs

qs.extend_pandas()

returns.sharpe()
```

```python
qs.plots.snapshot(returns=returns,
                  title='cs500 Performance')
```

```python
qs.stats.max_drawdown(stock)
```

```python
drawdown = qs.stats.to_drawdown_series(returns=returns)
display(drawdown)
```

```python
qs.stats.drawdown_details(drawdown=drawdown)
```

```python
qs.stats.outlier_win_ratio(returns=returns)
```

```python
qs.reports.full(returns=returns)
```

```python
qs.stats.volatility(returns=returns)
```

```python
qs.stats.rolling_volatility(returns=returns,
                            rolling_period=21,
                            periods_per_year=252)
```

```python
qs.plots.rolling_volatility(returns=returns,
                            period=21,
                            period_label='1-month',
                            periods_per_year=252,
                            fontname='Dengxian')
```
